```python
%%time
import numpy as np
import osmnx as ox

lat = 51.950296032982756
lng = 7.615756988525391

(51.95395646644701, 7.651804248986446)

# Get the OSM graph of "drive" streets for mnster an plot it
graph = ox.graph.graph_from_bbox(51.9925, 
    51.9045, 
    7.6891, 
    7.5352 , network_type='drive',simplify=True, retain_all=False)
#graph = ox.graph_from_point((lat, lng), dist=10000, network_type='drive')
ox.plot_graph(graph, fig_height=10, fig_width=10)
```

```python
from leuvenmapmatching.matcher.distance import DistanceMatcher
from leuvenmapmatching.matcher.newsonkrumm import NewsonKrummMatcher
from leuvenmapmatching.map.inmem import InMemMap

# Leuven Map Matching is using a different internal graph structure for the street data. 
# Therefore, the OSMnx graph needs to be transformed to the InMemMap
streetmap = InMemMap("enviroCar", use_latlon=True, use_rtree=True, index_edges=True)

# add nodes
lats = []
lons = []

nodes = list(graph.nodes)
for node in nodes:
    lng = graph.nodes[node]['x']
    lons.append(lng)
    lat = graph.nodes[node]['y']
    lats.append(lat)
    streetmap.add_node(node, (lat, lng))
    
    
# add edges
edges = list(graph.edges)
for edge in edges:
    node_a, node_b = edge[0], edge[1]
    streetmap.add_edge(node_a, node_b)
    
    # exclude bi-directional edges when street is oneway
    if not graph.edges[edge]['oneway']:
        streetmap.add_edge(node_b, node_a)
    
streetmap.purge()
```

```python
%%time
some_track_id = track_df['track.id'].unique()[2]
some_track = track_df[track_df['track.id'] == some_track_id]

some_track.plot()

#some_track = track_df 
#some_track.to_csv('track.csv')
```

```python
# Get the latitude and longitude coordinates from the track and join them in an array
lats = list(some_track['geometry'].apply(lambda coord: coord.y))
lngs = list(some_track['geometry'].apply(lambda coord: coord.x))

track_coords = [c for c in zip(lats, lngs)]

print(len(track_coords))

last_id = 0;
first_id = 0

max_y = max(lats) - .01
min_y = min(lats) + .01
max_x = max(lons) - .01
min_x = min(lons) + .01

for point in track_coords:
    if point[0] < min_y or point[1] > max_y or point[1] < min_x or point[1] > max_x:
        if(first_id != 0):
            break
    else:
        if(first_id == 0):
            first_id = last_id
        
    last_id = last_id +1


track_coords = track_coords[first_id:last_id]
#track_coords = track_coords[:176]

print(len(track_coords))
print(last_id)
    
```

```python
%%time

from pyproj import Geod


# Configure the mapmatcher. For increasing the performance it is important 
# that max_dist (in meters) is not set too high
matcher = NewsonKrummMatcher(streetmap,
                            max_dist=50, 
                            max_dist_init=100,
                            min_prob_norm=0.001,
                            non_emitting_length_factor=0.75,
                            obs_noise=50,
                            obs_noise_ne=75,
                            dist_noise=50,
                            non_emitting_edgeid=True)


# Perform the mapmatching 
edge_ids, last_idx = matcher.match(track_coords, unique=False)

# Reference ellipsoid for distance
geod = Geod(ellps='WGS84')
proj_dist = np.zeros(len(track_coords))

# edgeid refers to edges id (node1_id, node2_id) where the GPS point is projected
lat_corr, lon_corr = [], []
lat_nodes = matcher.lattice_best
for idx, m in enumerate(lat_nodes):
    if(idx == len(track_coords)):
        break;
    lat, lon = m.edge_m.pi[:2]
    lat_corr.append(lat)
    lon_corr.append(lon)
    # print(idx)
    _, _, distance = geod.inv(track_coords[idx][1], track_coords[idx][0], lon, lat)
    proj_dist[idx] += distance

    
track_corr = np.column_stack((lat_corr, lon_corr))
# Stack the stats
path_length = []
unlinked = []

print(track_corr)
print(lat_nodes)

# Compute the route coordinates
route = []

for i in range(len(track_corr) - 1):
    if edge_ids[i] != edge_ids[i+1]:
        route.append(track_corr[i])
        route.append([streetmap.graph[edge_ids[i][1]][0][0], streetmap.graph[edge_ids[i][1]][0][1]])
        _, _, distance = geod.inv(track_corr[i][1], track_corr[i][0],
                                      streetmap.graph[edge_ids[i][1]][0][1], streetmap.graph[edge_ids[i][1]][0][0])
        path_length.append(distance)
        unlinked.append(0)

    else:
        route.append(track_corr[i])
        _, _, distance = geod.inv(track_corr[i][1], track_corr[i][0],
                                      track_corr[i+1][1], track_corr[i+1][0])
        path_length.append(distance)
        unlinked.append(0)
# Let's not forget the last point
route.append(track_corr[-1])
route = np.array(route)

```

```python
def plot_html(track, track_corr=[],
             track_color="black", track_corr_color="#CD473E",
             track_size=2, track_corr_size=2,
             route_corr=[],
             route_size=2, route_corr_size=2,
             route_color="black", route_corr_color="#CD473E",
             route_opacity=.7, route_corr_opacity=.7,
             proj=False, proj_color="#CD473E", proj_size=1, proj_alpha=.5,
             show_graph=False, graph=None,
             file_name="my_map.html", save=True
             ):
    """
    

    Parameters
    ----------
    track : TYPE
        DESCRIPTION.
    track_corr : TYPE, optional
        DESCRIPTION. The default is [].
    track_color : TYPE, optional
        DESCRIPTION. The default is "black".
    track_corr_color : TYPE, optional
        DESCRIPTION. The default is "darkcyan".
    track_size : TYPE, optional
        DESCRIPTION. The default is 2.
    track_corr_size : TYPE, optional
        DESCRIPTION. The default is 2.
    route_corr : TYPE, optional
        DESCRIPTION. The default is [].
    route_size : TYPE, optional
        DESCRIPTION. The default is 2.
    route_corr_size : TYPE, optional
        DESCRIPTION. The default is 2.
    route_color : TYPE, optional
        DESCRIPTION. The default is "black".
    route_corr_color : TYPE, optional
        DESCRIPTION. The default is "darkcyan".
    route_opacity : TYPE, optional
        DESCRIPTION. The default is .6.
    route_corr_opacity : TYPE, optional
        DESCRIPTION. The default is .6.
    proj : TYPE, optional
        DESCRIPTION. The default is False.
    proj_color : TYPE, optional
        DESCRIPTION. The default is "skyblue".
    proj_size : TYPE, optional
        DESCRIPTION. The default is 1.
    proj_alpha : TYPE, optional
        DESCRIPTION. The default is 1.
    show_graph : TYPE, optional
        DESCRIPTION. The default is False.
    graph : TYPE, optional
        DESCRIPTION. The default is None.
    file_name : TYPE, optional
        DESCRIPTION. The default is "my_map.html".
    save : TYPE, optional
        DESCRIPTION. The default is True.

    Returns
    -------
    my_map : TYPE
        DESCRIPTION.

    """

    med_lat = track[len(track)//2][0]
    med_lon = track[len(track)//2][1]

    # Load map centred on central coordinates
    my_map = folium.Map(location=[med_lat, med_lon], zoom_start=13)


    # If the route is given in input, plot both (original and corrected)
    if len(route_corr) > 0:
        # add lines
        folium.PolyLine(track, color=route_color, weight=route_size, opacity=route_opacity).add_to(my_map)
        folium.PolyLine(route_corr, color=route_corr_color, weight=route_corr_size, opacity=route_corr_opacity).add_to(my_map)


    # add dots
    for i in range(len(track)):
        folium.CircleMarker(location=[track[i][0], track[i][1]],
                            radius=track_size,
                            weight=1,
                            color=track_color,
                            fill=True,
                            fill_opacity=1).add_to(my_map)
    for i in range(len(track_corr)):
        folium.CircleMarker(location=[track_corr[i][0], track_corr[i][1]],
                            radius=track_corr_size,
                            weight=1,
                            color=track_corr_color,
                            fill=True,
                            fill_opacity=1).add_to(my_map)


    # add the OSM light grey background
    folium.TileLayer('cartodbpositron').add_to(my_map)

    # plot the legend in the HTML page
    legend_html = """
    <div style="position: fixed;
                width: 210px;
                top: 10px; right: 10px;
                border: 2px solid lightgrey;
                border-radius: 4px;
                background-color: rgba(255, 255, 255, 0.85);
                z-index:9999;
                font-size: 15px; color: slategrey;
     ">
         &nbsp; <span style="font-weight: bold">Legend</span>
         <br>
             &nbsp; Original Point &nbsp;
             <i class="fa fa-circle"
                 style="float: right;
                         margin-right: 19px; margin-top: 4px;
                         color: black">
             </i>
         <br>
             &nbsp; Projected Point &nbsp;
             <i class="fa fa-circle"
                 style="float: right;
                         margin-right: 19px; margin-top: 4px;
                         color: #CD473E">
             </i>
         <br>
             &nbsp; Projection &nbsp;
             <div class="line"
                 style="float: right;
                         margin-right: 10px; margin-top: 10px;
                         width: 30px; height: 2px;
                         background-color: #CD473E">
             </div>
    </div>
    """
    my_map.get_root().html.add_child(folium.Element(legend_html))

    if save:
        my_map.save(file_name)
        # Plot in new tab
        #webbrowser.open(file_name, new=2)  # open in new tab

    return my_map
```

```python
edgeset = some_track
import osmapi as osm
api = osm.OsmApi() 

edge_agg = edgeset.groupby(['st_nodes', 'end_nodes']).agg({'Speed.value': ['mean', 'min', 'max']})
edgeset = pd.merge(edgeset, edge_agg,  how='left', left_on=['st_nodes','end_nodes'], right_on = ['st_nodes','end_nodes'])
#edgeset = edgeset.drop(['randomSpeed'], axis=1).drop_duplicates()
edgeset

result_array = np.array([])
arrStartNodelat = np.array([])
arrStartNodelon = np.array([])
arrEndNodelat = np.array([])
arrEndNodelon = np.array([])

st_rows = edgeset['st_nodes']
end_rows = edgeset['end_nodes']


for row in zip(st_rows, end_rows):
    stNode = api.NodeGet(row[0])
    arrStartNodelat = np.append(arrStartNodelat, stNode["lat"])
    arrStartNodelon = np.append(arrStartNodelon, stNode["lon"])
    
    endNode = api.NodeGet(row[1])
    arrEndNodelat = np.append(arrEndNodelat,endNode["lat"])
    arrEndNodelon = np.append(arrEndNodelon, endNode["lon"])
   
arrStartNodelat= pd.DataFrame(arrStartNodelat)
arrStartNodelon= pd.DataFrame(arrStartNodelon)
arrEndNodelat= pd.DataFrame(arrEndNodelat)
arrEndNodelon= pd.DataFrame(arrEndNodelon)


edgeset['StNode_lat']=arrStartNodelat
edgeset['StNode_lon']=arrStartNodelon
edgeset['EndNode_lat']=arrEndNodelat
edgeset['EndNode_lon']=arrEndNodelon

edgeset
```
